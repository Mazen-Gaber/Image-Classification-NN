# **Loading CIFAR-10 Dataset**

```python
transform = transforms.ToTensor()
```

```python
data_path = '../data_cifar/'
cifar10_train = datasets.CIFAR10(data_path, train=True, download=True, transform=transform)
cifar10_test = datasets.CIFAR10(data_path, train=False, download=True, transform=transform)
```

```python
torch.manual_seed(80)
train_loader = DataLoader(cifar10_train, batch_size=32, shuffle=True)
test_loader = DataLoader(cifar10_test, batch_size=100, shuffle=False)
```

## **Displaying the batch**

```python
dataiter = iter(train_loader)
images, labels = next(dataiter)
viz = torchvision.utils.make_grid(images, nrow=8, padding = 1).numpy()
fig, ax = plt.subplots(figsize= (8,8))
ax.imshow(np.transpose(viz, (1,2,0)))
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
dataiter = iter(test_loader)
images, labels = next(dataiter)
viz = torchvision.utils.make_grid(images, nrow=10, padding = 1).numpy()
fig, ax = plt.subplots(figsize= (8,8))
ax.imshow(np.transpose(viz, (1,2,0)))
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```
